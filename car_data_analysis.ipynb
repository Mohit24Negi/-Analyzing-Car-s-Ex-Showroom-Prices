```python
def convert_torque(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower()
    x = " ".join(x.split())
    x = x.replace(" ", "")

    if "kgm" in x:
        return float(x.split("kgm")[0]) * 9.81
    elif "nm" in x:
        return float(x.split("nm")[0])
    elif "@" in x:
        return float(x.split("@")[0])
    else:
        return float(x)  

ob_df["Torque"] = ob_df["Torque"].apply(convert_torque)


```

* From **Number_of_Airbags vs Car Price** we can observe a **positive correlation.**
* **Infernece:** With **increase** in **number of airbages in car** the **price of car increases.**
