## 01. Cleaning Make and Model column.

### Displaying unique values in Make.

```python
print("unique values of Make column: \n\n",ob_df["Make"].unique())
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="left"><font color=brown>📊 Inference:</font></h3>


* There are some **typographical error / duplicate values** in list of **Make Feature.**
    1. **`Land Rover Rover`** should be **`Land Rover`.**
 
    2. **`Maruti Suzuki R`** should be **`Maruti Suzuki`.**
 
    3. **`Dc`** should be **`DC`.**
 
    4. **`Bmw`** should be **`BMW`.**

```python
ob_df['Make'] = ob_df['Make'].replace({
    'Land Rover Rover': 'Land Rover',
    'Maruti Suzuki R': 'Maruti Suzuki',
    'Dc': 'DC',
    'Bmw': 'BMW',
    'Mg': 'MG',
    'Icml': 'ICML'
})
```

```python
ob_df[ob_df["Make"].isna()][["Make", "Model"]]
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="left"><font color=brown>📊 Inference:</font></h3>

- We observe that when the **Make** feature has **missing values**, the **Model** feature often contains the **car brand at the beginning**.
- Therefore, we can **extract the first word** from the **Model** column and use it to **fill the missing values** in the **Make** feature.
- However, before applying this logic, we need to **standardize the data**. Specifically, we should replace **"Go+"** with **"Datsun GO+"** in the **Model** column.
- This ensures that we can **consistently extract the car brand** from the **Model** column using a **single, unified approach**.

```python
ob_df["Model"] = ob_df["Model"].replace({"Go+" : "Datsun Go+"})
```

```python
mask1 = ob_df["Make"].isna()
ob_df.loc[mask1, "Make"] = ob_df.loc[mask1, "Model"].str.split().str[0]

mask2 = ob_df["Make"] == ob_df["Model"].str.split().str[0]
ob_df.loc[mask2, "Model"] = ob_df.loc[mask2, "Model"].str.split().str[1:].str.join(" ")
```

```python
print("Total Number of missing value:-", ob_df["Make"].isna().sum())
```

<b>Body_Type:</b> Has duplicate swapped combos (<i>SUV, Crossover</i> vs <i>Crossover, SUV</i>), multi-label entries, non-standard <i>Sports</i> category, and missing values — simplify to one primary body type per row.<br><br>
<b>Emission_Norm:</b> <i>BS 6</i> and <i>BS VI</i> represent the same standard but are treated as two different categories — standardize all to Roman numerals.


```python
body_map = {
    'SUV, Crossover'      : 'SUV',
    'Crossover, SUV'      : 'SUV',
    'Sedan, Coupe'        : 'Sedan',
    'Sedan, Crossover'    : 'Sedan',
    'Sports, Hatchback'   : 'Hatchback',
    'Sports, Convertible' : 'Convertible',
    'Coupe, Convertible'  : 'Convertible',
    'Sports'              : 'Coupe',
}
ob_df['Body_Type'] = ob_df['Body_Type'].replace(body_map)


ob_df['Emission_Norm'] = ob_df['Emission_Norm'].replace({
    'BS 6' : 'BS VI',
    'BS 4' : 'BS IV',
    'BS 3' : 'BS III',
})
```

- We can see that **Displacement** and **Torque** contain **missing values**. We can fill these missing values using the **average (mean)** of each feature.

### Extracting Numerical values.

```python
# Extract numerical values from Displacement column

ob_df['Displacement'] = (
    ob_df['Displacement']
    .str.extract(r'(\d+\.?\d*)')   # Extract numbers
    .astype(float)                 # Convert to float
)

# Fill missing values with mean
ob_df['Displacement'].fillna(
    ob_df['Displacement'].mean(),
    inplace=True
)

ob_df['Displacement'] = ob_df['Displacement'].astype(int)
```

```python
mask = ob_df['Power'].astype(str).str.contains('nm', case=False, na=False)

# Swap Power and Torque values
ob_df.loc[mask, ['Power', 'Torque']] = (
    ob_df.loc[mask, ['Torque', 'Power']].values
)
```

```python
# Function to convert all power units into HP

def convert_to_hp(value):

    if pd.isna(value):
        return np.nan

    value = str(value).strip()

    # Split using @ and take left side
    power_part = value.split('@')[0].strip().lower()

    # Extract numeric value without using re
    num = ''.join(ch for ch in power_part if ch.isdigit() or ch == '.')

    if num == '':
        return np.nan

    num = float(num)

    # Convert units into HP
    if 'ps' in power_part:
        return round(num * 0.98632, 2)   # PS → HP

    elif 'kw' in power_part:
        return round(num * 1.34102, 2)   # kW → HP

    else:
        return round(num, 2)             # hp / bhp already in HP


# Apply conversion
ob_df['Power'] = ob_df['Power'].apply(convert_to_hp)

# Fill missing values with mean
ob_df['Power'].fillna(
    ob_df['Power'].mean(),
    inplace=True
)
```

```python
# Function to convert Torque values into Nm

def convert_to_nm(value):

    if pd.isna(value):
        return np.nan

    value = str(value).strip().lower()

    # Split using @ and take left side
    torque_part = value.split('@')[0].strip()

    # Extract numeric value
    num = ''.join(ch for ch in torque_part if ch.isdigit() or ch == '.')

    if num == '':
        return np.nan

    num = float(num)

    # Convert units into Nm
    if 'kgm' in torque_part:
        return round(num * 9.80665, 2)   # kgm → Nm

    else:
        return round(num, 2)             # Nm already


# Apply conversion
ob_df['Torque'] = ob_df['Torque'].apply(convert_to_nm)

# Fill missing values with mean
ob_df['Torque'].fillna(
    ob_df['Torque'].mean(),
    inplace=True
)
```
